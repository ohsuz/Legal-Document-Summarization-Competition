### **등장횟수 기반 딕셔너리**

```python
from collections import Counter, defaultdict
```

```python
count_dict = {}

for i in range(len(a1)):
  row = [a1[i], a2[i], a3[i], b1[i], b2[i], b3[i], c1[i], c2[i], c3[i], d1[i], d2[i], d3[i], e1[i], e2[i], e3[i],
         f1[i], f2[i], f3[i], g1[i], g2[i], g3[i], h1[i], h2[i], h3[i], i1[i], i2[i], i3[i], j1[i], j2[i], j3[i],
         k1[i], k2[i], k3[i], l1[i], l2[i], l3[i], m1[i], m2[i], m3[i], n1[i], n2[i], n3[i], o1[i], o2[i], o3[i],
         p1[i], p2[i], p3[i], q1[i], q2[i], q3[i]]
  count_dict[ID[i]] = [tup[0] for tup in Counter(row).most_common()][:3]
```

```python
print(count_dict)
```

```python
count_1, count_2, count_3 = [], [], []

for values in list(count_dict.values()):
  count_1.append(values[0])
  count_2.append(values[1])
  count_3.append(values[2])
```

### **중요도 기반 딕셔너리**

```python
importance_dict = defaultdict(int)

for i in range(len(a1)):
  score_dict = defaultdict(int)
  row1 = [a1[i], b1[i], c1[i], d1[i], e1[i]]
  row2 = [a2[i], b2[i], c2[i], d2[i], e2[i]]
  row3 = [a3[i], b3[i], c3[i], d3[i], e3[i]]

  for row in row1:
    score_dict[row] += 3
  for row in row2:
    score_dict[row] += 2
  for row in row3:
    score_dict[row] += 1

  importance_dict[ID[i]] = [tup[0] for tup in Counter(score_dict).most_common()][:3]
```

```python
print(importance_dict)
```

```python
importance_1, importance_2, importance_3 = [], [], []

for values in list(importance_dict.values()):
  importance_1.append(values[0])
  importance_2.append(values[1])
  importance_3.append(values[2])
```

### **최종 Dataframe**

```python
final_df = pd.DataFrame(data=ID, columns=['ID'])

final_df['count_1'] = count_1
final_df['count_2'] = count_2
final_df['count_3'] = count_3
final_df['importance_1'] = importance_1
final_df['importance_2'] = importance_2
final_df['importance_3'] = importance_3
```
